### 1)SAMSUNG Galaxy Watch Ultra LTE  (Titanium Gray Strap, Large)

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import pandas as pd
import time
import random
from gender_guesser.detector import Detector

detector = Detector()

CHROME_DRIVER_PATH = "C:\\Users\\tanwa\\Downloads\\chromedriver-win64\\chromedriver-win64\\chromedriver.exe" 

options = Options()
options.add_argument("--headless") 
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("--window-size=1920x1080")
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36")

service = Service(CHROME_DRIVER_PATH)
driver = webdriver.Chrome(service=service, options=options)

try:
    url = "https://www.flipkart.com/samsung-galaxy-watch-ultra-lte/product-reviews/itmc66e51faa52df?pid=SMWH2EHFNBFCBXFK&lid=LSTSMWH2EHFNBFCBXFKVBUCWQ&marketplace=FLIPKART"
    driver.get(url)
    time.sleep(5) 
    
    for _ in range(3):  
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
        time.sleep(2)

    reviews = driver.find_elements(By.XPATH, '//div[@class="ZmyHeo"]') 
    usernames = driver.find_elements(By.XPATH, '//p[@class="_2NsDsF AwS1CA"]')
    locations = driver.find_elements(By.XPATH, '//p[@class="MztJPv"]')
    
    reviews_list1 = [review.text for review in reviews]
    usernames_list1 = [username.text for username in usernames]
    locations_list1 = [location.text for location in locations]

    genders = [detector.get_gender(name.split()[0]) for name in usernames_list1]

    age_groups = random.choices(
        ["Under 18", "18-24", "25-34", "35-44", "45-54", "55 and above"],
        k=len(reviews_list1)
    )

    occupations = random.choices(
        ["Student", "Working Professional", "Homemaker", "Retired", "Freelancer"],
        k=len(reviews_list1)
    )

    print("Reviews:")
    for i in range(len(reviews_list1)):
        review = reviews_list1[i]
        gender = genders[i]
        print(review)
        print(gender)
        print(f"Age : {age_groups[i]}")
        print(f"Location : {locations_list1[i]}")
        print(f"Occupation: {occupations[i]}")
        print("-" * 80)
    if reviews_list1:
        df = pd.DataFrame({
            "Reviews for Samsung Galaxy Watch Ultra": reviews_list1,
            "Gender":genders,
            "Location":locations_list1,
            "Occupation":occupations,
            "Age":age_groups
        })

        df.to_excel("flipkart_reviews_with_gender_location_occupation1.xlsx", index=False, engine="openpyxl")

        print("Reviews saved successfully to 'flipkart_reviews_with_gender_location_occupation1.xlsx'!")

    else:
        print("No reviews found. Please check the XPath or page structure.")
except Exception as e:
    print(f"An error occurred: {e}")

finally:
    driver.quit()
```
