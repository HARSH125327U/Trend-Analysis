```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "C:\\Users\\tanwa\\Downloads\\Colab Notebooks-20241019T120126Z-001\\Colab Notebooks\\Trend Analysis\\customer_reviews_with_sentiment.xlsx"  # Replace with your file path
data = pd.read_excel(file_path)

# Replace these with actual column names in your dataset
sentiment_column = "Sentiment"
age_column = "Age"
location_column = "Location"
occupation_column = "Occupation"

# 1. Total reviews count
total_reviews = len(data)
print(f"Total Reviews: {total_reviews}")

# 2. Reviews by Age Group
reviews_by_age_group = data[age_column].value_counts()
print("Reviews by Age Group:")
print(reviews_by_age_group)

# 3. Reviews by Location
reviews_by_location = data[location_column].value_counts()
print("Reviews by Location:")
print(reviews_by_location)

# 4. Reviews by Occupation
reviews_by_occupation = data[occupation_column].value_counts()
print("Reviews by Occupation:")
print(reviews_by_occupation)

# 5. Reviews by Sentiment
reviews_by_sentiment = data[sentiment_column].value_counts()
print("Reviews by Sentiment:")
print(reviews_by_sentiment)

# Visualization: Reviews by Age Group
plt.figure(figsize=(8, 6))
reviews_by_age_group.plot(kind='bar', title="Reviews by Age Group", ylabel="Number of Reviews", xlabel="Age Group")
plt.show()

# Visualization: Reviews by Location
plt.figure(figsize=(8, 6))
reviews_by_location.plot(kind='bar', title="Reviews by Location", ylabel="Number of Reviews", xlabel="Location")
plt.show()

# Visualization: Reviews by Sentiment
plt.figure(figsize=(8, 6))
reviews_by_sentiment.plot(kind='pie', autopct='%1.1f%%', title="Reviews by Sentiment")
plt.ylabel('')  # Remove y-axis label for a cleaner pie chart
plt.show()

# Optional: Save the analysis to an Excel file
output_file = "usage_analysis.xlsx"
analysis_summary = {
    "Reviews by Age Group": reviews_by_age_group,
    "Reviews by Location": reviews_by_location,
    "Reviews by Occupation": reviews_by_occupation,
    "Reviews by Sentiment": reviews_by_sentiment,
}
summary_df = pd.concat(analysis_summary, axis=1)
summary_df.to_excel(output_file)

print(f"Usage analysis saved to {output_file}")
```
