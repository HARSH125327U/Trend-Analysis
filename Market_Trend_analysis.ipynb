```python
import pandas as pd
import numpy as np
from datetime import datetime

# Load your file
file_path = "C:\\Users\\tanwa\\Downloads\\Final Trend analysis(final renew).xlsx"
df = pd.read_excel(file_path, sheet_name="Sheet1")

# Define seasons with their months
seasons = {
    "Winter": ["December", "January", "February"],
    "Spring": ["March", "April", "May"],
    "Summer": ["June", "July", "August"],
    "Autumn": ["September", "October", "November"]
}

# Generate random Month and Year simultaneously
years = list(range(2020, 2025))
df["Season"] = np.random.choice(list(seasons.keys()), len(df))
df["Month"] = df["Season"].apply(lambda s: np.random.choice(seasons[s]))
df["Year"] = np.random.choice(years, len(df))

# Save the updated file
df.to_excel("C:\\Users\\tanwa\\Downloads\\Final Trend analysis(final renew).xlsx", sheet_name="Sheet1", index=False)
```
