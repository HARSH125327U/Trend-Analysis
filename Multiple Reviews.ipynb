```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import pandas as pd
import time
import random
from gender_guesser.detector import Detector

detector = Detector()

CHROME_DRIVER_PATH = "C:\\Users\\tanwa\\Downloads\\chromedriver-win64\\chromedriver-win64\\chromedriver.exe" 

options = Options()
options.add_argument("--headless") 
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("--window-size=1920x1080")
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36")

service = Service(CHROME_DRIVER_PATH)
driver = webdriver.Chrome(service=service, options=options)

urls = [
    "https://www.flipkart.com/fossil-4th-gen-explorist-hr-smartwatch/product-reviews/itm82f6c904e2140?pid=SMWFBDGYBGUHYFMF&lid=LSTSMWFBDGYBGUHYFMFGYKHZG&marketplace=FLIPKART",
    "https://www.flipkart.com/fossil-4th-gen-explorist-hr-smartwatch/product-reviews/itm82f6c904e2140?pid=SMWFBDGYBGUHYFMF&lid=LSTSMWFBDGYBGUHYFMFGYKHZG&marketplace=FLIPKART&page=2",
    "https://www.flipkart.com/fossil-4th-gen-explorist-hr-smartwatch/product-reviews/itm82f6c904e2140?pid=SMWFBDGYBGUHYFMF&lid=LSTSMWFBDGYBGUHYFMFGYKHZG&marketplace=FLIPKART&page=3",
    "https://www.flipkart.com/fossil-4th-gen-explorist-hr-smartwatch/product-reviews/itm82f6c904e2140?pid=SMWFBDGYBGUHYFMF&lid=LSTSMWFBDGYBGUHYFMFGYKHZG&marketplace=FLIPKART&page=4",
    "https://www.flipkart.com/fossil-4th-gen-explorist-hr-smartwatch/product-reviews/itm82f6c904e2140?pid=SMWFBDGYBGUHYFMF&lid=LSTSMWFBDGYBGUHYFMFGYKHZG&marketplace=FLIPKART&page=5",
    "https://www.flipkart.com/fossil-4th-gen-explorist-hr-smartwatch/product-reviews/itm82f6c904e2140?pid=SMWFBDGYBGUHYFMF&lid=LSTSMWFBDGYBGUHYFMFGYKHZG&marketplace=FLIPKART&page=6",
    "https://www.flipkart.com/fossil-4th-gen-explorist-hr-smartwatch/product-reviews/itm82f6c904e2140?pid=SMWFBDGYBGUHYFMF&lid=LSTSMWFBDGYBGUHYFMFGYKHZG&marketplace=FLIPKART&page=7",
    "https://www.flipkart.com/fossil-4th-gen-explorist-hr-smartwatch/product-reviews/itm82f6c904e2140?pid=SMWFBDGYBGUHYFMF&lid=LSTSMWFBDGYBGUHYFMFGYKHZG&marketplace=FLIPKART&page=8",
    "https://www.flipkart.com/fossil-4th-gen-explorist-hr-smartwatch/product-reviews/itm82f6c904e2140?pid=SMWFBDGYBGUHYFMF&lid=LSTSMWFBDGYBGUHYFMFGYKHZG&marketplace=FLIPKART&page=9",
    "https://www.flipkart.com/fossil-4th-gen-explorist-hr-smartwatch/product-reviews/itm82f6c904e2140?pid=SMWFBDGYBGUHYFMF&lid=LSTSMWFBDGYBGUHYFMFGYKHZG&marketplace=FLIPKART&page=10"
]
all_reviews = []
all_usernames = []
all_locations = []
try:
    for url in urls:
        print(f"Scraping: {url}")
        driver.get(url)
        time.sleep(5) 
        for _ in range(3):  
            driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
            time.sleep(2)

        reviews = driver.find_elements(By.XPATH, '//div[@class="ZmyHeo"]') 
        usernames = driver.find_elements(By.XPATH, '//p[@class="_2NsDsF AwS1CA"]')
        locations = driver.find_elements(By.XPATH, '//p[@class="MztJPv"]')
    
        all_reviews.extend([review.text for review in reviews])
        all_usernames.extend([username.text for username in usernames])
        all_locations.extend([location.text for location in locations])

        genders = [detector.get_gender(name.split()[0]) for name in all_usernames]

        age_groups = random.choices(
            ["Under 18", "18-24", "25-34", "35-44", "45-54", "55 and above"],
            k=len(all_reviews)
        )

        occupations = random.choices(
            ["Student", "Working Professional", "Homemaker", "Retired", "Freelancer"],
            k=len(all_reviews)
        )
        
        print("All Reviews:\n")
        for i, review in enumerate(all_reviews, start=1):
            print(f"Review {i}: {review}")
            print("-" * 80)

        if all_reviews:
            df = pd.DataFrame({
                "Reviews for Fossil 4th Gen Explorist HR": all_reviews,
                "Age":age_groups,
                "Gender":genders,
                "Location":all_locations,
                "Occupation":occupations
            })

            df.to_excel("a80.xlsx", index=False, engine="openpyxl")

            print("Reviews saved successfully to 'a80.xlsx'!")

        else:
            print("No reviews found. Please check the XPath or page structure.")
except Exception as e:
    print(f"An error occurred: {e}")

finally:
    driver.quit()
```
